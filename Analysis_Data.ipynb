## 7. Apakah usia saat masuk perguruan tinggi memengaruhi kemungkinan mahasiswa untuk lulus?

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='Target', y='Age at enrollment')
plt.title('Usia Saat Masuk vs Status Akhir Mahasiswa')
plt.xlabel('Status Akhir')
plt.ylabel('Usia Saat Masuk')
plt.tight_layout()
plt.show()

# T-test usia
graduate_age = df[df['Target'] == 'Graduate']['Age at enrollment']
dropout_age = df[df['Target'] == 'Dropout']['Age at enrollment']
t_stat, p_val = ttest_ind(graduate_age, dropout_age, nan_policy='omit')
print(f'T-Test Usia Masuk: t={t_stat:.2f}, p={p_val:.4f}')
```

Grafik box plot menunjukkan bahwa usia saat masuk perguruan tinggi tidak memiliki perbedaan yang signifikan secara visual pada tingkat kelulusan mahasiswa. Median usia di antara kelompok dropout, lulus, dan masih terdaftar relatif sama. Meskipun terdapat beberapa outlier, rentang antar kuartil (IQR) juga relatif konsisten di ketiga kelompok. Oleh karena itu, berdasarkan visualisasi ini, usia saat masuk perguruan tinggi tampaknya tidak menjadi faktor prediktor kuat untuk keberhasilan kelulusan. namun hal ini perlu di telaah lebih lagi, dikarenakan rentang usia yang lulus itu dibawah 20 hingga 20 keatas saja, sedangkan yang tidak lulus bisa lebih dari umur 25 hingga 30. dalam prespektif usia, jarak 2-3 tahun saja bianya cukup mempengaruhi kualitas belajar.

# Data Preparation / Preprocessing

- Mengelompokkan fitur yang kemungkinan terdapat outlier

```python
outlier_features = [
    'Curricular units 1st sem (enrolled)',
    'Curricular units 1st sem (evaluations)', 'Curricular units 1st sem (approved)', 
    'Curricular units 1st sem (grade)', 'Curricular units 2nd sem (credited)', 
    'Curricular units 2nd sem (enrolled)', 'Curricular units 2nd sem (evaluations)', 
    'Curricular units 2nd sem (approved)', 'Curricular units 2nd sem (grade)'
]
```

- Menggunakan boxplot untuk memudahkan deteksi outlier

```python
# Boxplot untuk visualisasi outliers
plt.figure(figsize=(14, 10))
for i, col in enumerate(outlier_features):
    plt.subplot(3, 3, i+1)
    sns.boxplot(x=df[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```

- Menghitung z-core untuk deteksi jumlah outlier yang harus ditangani

```python
# Menghitung z-score untuk deteksi outlier
z_scores = np.abs(zscore(df[outlier_features]))
outliers_mask = (z_scores > 3)
print("Jumlah outliers per kolom:")
print(pd.DataFrame(outliers_mask, columns=outlier_features).sum())
```

terdapat beberapa outlier, disini akan melakuan deteksi baris mana saja yang memiliki minimal 1 outlier

```python
rows_with_outliers = (outliers_mask.sum(axis=1) > 0)
print(f"Jumlah baris dengan outlier: {rows_with_outliers.sum()}")
```

- Mengambil hanya data yang tidak ada Outlier

```python
# Buang baris dengan outlier
df_clean = df[~rows_with_outliers].copy()
```

- Mengecek ulang apakah masih ada Oulier

```python
# Boxplot untuk visualisasi outliers
plt.figure(figsize=(14, 10))
for i, col in enumerate(outlier_features):
    plt.subplot(3, 3, i+1)
    sns.boxplot(x=df_clean[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```

- Capping outlier (winsorizing): mengganti nilai outlier dengan batas atas/bawah.

```python
for col in outlier_features:
    Q1 = df_clean[col].quantile(0.25)
    Q3 = df_clean[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df_clean[col] = np.where(df_clean[col] > upper, upper,
                             np.where(df_clean[col] < lower, lower, df_clean[col]))
```

- Mengecek kembali data apakah sudah bersih dari Outlier

```python
# Boxplot untuk visualisasi outliers
plt.figure(figsize=(14, 10))
for i, col in enumerate(outlier_features):
    plt.subplot(3, 3, i+1)
    sns.boxplot(x=df_clean[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```

- Menentukan features yang akan digunakan berdasarkan analisis eda yang sudah dilakukan

```python
features = [
    'Marital Status', 'Course', 'Daytime/evening attendance', 'Scholarship holder',
    'Age at enrollment', 'Curricular units 1st sem (credited)', 'Curricular units 1st sem (enrolled)',
    'Curricular units 1st sem (evaluations)', 'Curricular units 1st sem (approved)', 'Curricular units 1st sem (grade)',
    'Curricular units 2nd sem (credited)', 'Curricular units 2nd sem (enrolled)',
    'Curricular units 2nd sem (evaluations)', 'Curricular units 2nd sem (approved)', 'Curricular units 2nd sem (grade)'
]
```

- Membagi X sebagai features dan Y sebagai Target

```python
X = df_clean[features]
y = df_clean['Target']
```

- Dilakukan pengecekan distribusi data sebelum dilakukan undersampling

```python
print("Distribusi sebelum balancing:", Counter(y))
```

- melakukan under sampling pada data

```python
rus = RandomUnderSampler(sampling_strategy='not minority', random_state=42)
X_balanced, y_balanced = rus.fit_resample(X, y)

print("Distribusi setelah undersampling:", Counter(y_balanced))
```

- dilakukan encoding pada kolom target dan scalling pada kolom features sebelum dilakukan modeling 

```python
# Encode target jika masih dalam bentuk string
# 'Graduate', 'Dropout', 'Enrolled' -> encode jadi 0/1/2
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y_balanced)
```

```python
# Feature Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_balanced)
```

- Membagi data sebagai Train dan test sebelum dilakukan modeling

```python
# Split data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_encoded, test_size=0.2, random_state=42, stratify=y_balanced)
```

## Modelling

- Model dengan menggunakan Random Forest

```python
# Train Model
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
```

- Model dengan menggunakan Logistic Regression
  

```python
lr_model = LogisticRegression(max_iter=1000, multi_class='multinomial', solver='lbfgs')
lr_model.fit(X_train, y_train)
```

## Evaluation 

- Evaluasi Matrix pada model Random Forest

```python
# Evaluasi Random Forest Classifier
y_pred = model.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred, target_names=label_encoder.classes_))
```

- Evaluasi Matrix pada model Random Forest

```python
# Evaluasi Logistic Regression
y_pred = lr_model.predict(X_test)

print("Classification Report:\n", classification_report(y_test, y_pred, target_names=label_encoder.classes_))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
